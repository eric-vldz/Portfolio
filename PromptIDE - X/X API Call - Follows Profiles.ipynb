```python
def xmetrics(*usernames):
    """
    Fetch user data and their most recent tweet for multiple usernames from X's API.

    Args:
    *usernames: Variable number of usernames to query.

    Returns:
    DataFrame: A pandas DataFrame containing user data and their most recent tweet if available.
    """

    # User fields to pull from X API
    user_fields = "affiliation,connection_status,created_at,description,entities,id,location,most_recent_tweet_id,name,pinned_tweet_id,profile_banner_url,profile_image_url,protected,public_metrics,receives_your_dm,subscription_type,url,username,verified,verified_type,withheld"
    
    # Tweet fields to pull from X API if fetching the most recent tweet
    tweet_fields = "article,attachments,author_id,card_uri,context_annotations,conversation_id,created_at,edit_controls,edit_history_tweet_ids,entities,geo,id,in_reply_to_user_id,lang,non_public_metrics,note_tweet,organic_metrics,possibly_sensitive,promoted_metrics,public_metrics,referenced_tweets,reply_settings,scopes,source,text,username,withheld"

    # Ensure authentication details are provided
    if not all([config['API_KEY'], config['API_SECRET_KEY'], config['ACCESS_TOKEN'], config['ACCESS_TOKEN_SECRET']]):
        raise ValueError("All authentication details must be provided in the config file.")

    auth = OAuth1(config['API_KEY'], config['API_SECRET_KEY'], config['ACCESS_TOKEN'], config['ACCESS_TOKEN_SECRET'])

    all_users_data = []

    for username in usernames:
        # Construct the URL with parameters for user data
        user_url = f'https://api.twitter.com/2/users/by/username/{username}?user.fields={user_fields}'

        # Making the GET request for user data
        user_response = requests.get(user_url, auth=auth)

        if user_response.status_code == 200:
            user_data = user_response.json()
            if 'data' in user_data:
                user_info = user_data['data']
                
                # If there's a most recent tweet, fetch it
                if 'most_recent_tweet_id' in user_info and user_info['most_recent_tweet_id']:
                    tweet_id = user_info['most_recent_tweet_id']
                    tweet_url = f'https://api.twitter.com/2/tweets/{tweet_id}?tweet.fields={tweet_fields}'
                    tweet_response = requests.get(tweet_url, auth=auth)
                    
                    if tweet_response.status_code == 200:
                        tweet_data = tweet_response.json()
                        if 'data' in tweet_data:
                            # Combine user data with tweet data
                            user_info['most_recent_tweet'] = tweet_data['data']
                        else:
                            print(f"No tweet data found for {username}")
                    else:
                        print(f"Failed to get tweet data for {username}. Status code: {tweet_response.status_code}")
                
                all_users_data.append(user_info)
            else:
                print(f"No 'data' key found for {username}")
        else:
            print(f"Failed to get data for user {username}. Status code: {user_response.status_code}")
            print(user_response.text)

    if all_users_data:
        df = pd.DataFrame(all_users_data)
        
        # Unnest public_metrics if present
        if 'public_metrics' in df.columns:
            public_metrics_df = pd.json_normalize(df['public_metrics'])
            public_metrics_df.columns = ['public_metrics.' + col for col in public_metrics_df.columns]
            df_expanded = pd.concat([df.drop('public_metrics', axis=1), public_metrics_df], axis=1)
            
            # If there's tweet data, unnest that too
            if 'most_recent_tweet' in df_expanded.columns:
                tweet_df = pd.json_normalize(df_expanded['most_recent_tweet'])
                tweet_df.columns = ['tweet.' + col for col in tweet_df.columns]
                df_expanded = pd.concat([df_expanded.drop('most_recent_tweet', axis=1), tweet_df], axis=1)
            
            return df_expanded
        else:
            return df

    return None

# Example usage:
df_expanded2 = xmetrics('InvestorVideos', 'TeloPongoTuro', 'TateAllen24')

if df_expanded2 is not None:
    print(df_expanded2.head())
```
