```r
# load packages and dataset
library(ISLR)
library(ggplot2)
library(cluster)
```

```r
# Create scatterplot of Petal Length vs. Petal Width
ggplot(iris, aes(Petal.Length, Petal.Width, color = Species)) + geom_point(size=4) + theme_bw()
```

```r
# Create K means cluster model
set.seed(101)
iris.cluster <- kmeans(iris[, 1:4], 3, nstart = 20)
print(iris.cluster)
table(iris.cluster$cluster, iris$Species)
```

```r
# Use clusplot to plot clusters
clusplot(iris, iris.cluster$cluster, color=TRUE, shade=TRUE, labels=0, lines=0)
```
